```python
import parser
import os
from glob import glob
import pandas as pd
from functools import reduce 
from xml.etree import ElementTree as et
```

```python
import glob

# Specify the directory where your XML files are located
xml_directory = './dataset/'

# Use glob to get a list of all XML files in the directory
xml_list = glob.glob(f'{xml_directory}/*.xml')
xml_list = list(map(lambda x: x.replace('\\', '/'), xml_list))

# Print the list of XML files
print("List of XML files:")
for xml_file in xml_list:
    print(xml_file)
```

```python

def extract_text(filename):
    tree= et.parse(filename)
    root=tree.getroot()
    width= root.find('size').find('width').text
    height= root.find('size').find('height').text
    image_name=root.find('filename').text
    print(height)
    objs= root.findall('object')
    parser=[]
    for obj in objs:

        name=obj.find('name').text
        bndbox=obj.find('bndbox')
        
        xmin=bndbox.find('xmin').text
        xmax=bndbox.find('xmax').text
        ymin=bndbox.find('ymin').text
        ymax=bndbox.find('ymax').text
        parser.append([image_name,width,height, name,xmin,xmax,ymin,ymax])

    return parser
```

```python
parser_all=list(map(extract_text,xml_list))
```

```python
data= reduce(lambda x,y : x+y,parser_all)
```

```python
def label_encoding(x):
    if x == 'menu':
        return 1
    elif x == 'menu bar':
        return 2
    elif x == 'menu-icon':
        return 3
    elif x == 'decorators':
        return 4
    else:
        return 0

```

```python
import os
from shutil import move
import pandas as pd
from sklearn.model_selection import train_test_split

dataset_folder = '/home/lenovo/temp-image-detec/img_data'
train_folder = os.path.join(dataset_folder, 'train')
test_folder = os.path.join(dataset_folder, 'test')

# Assuming you have already defined group_by_obj_train and group_by_obj_test

def save_data(filename, folder_path, group_obj):
    src = os.path.join(dataset_folder, filename)
    dst = os.path.join(folder_path, filename)

    # Check if the source file exists before moving
    if os.path.exists(src):
        move(src, dst)

        # Fixing the syntax error here
        text_filename = os.path.join(folder_path, os.path.splitext(filename)[0] + '.txt')

        group_obj.get_group(filename).set_index('filename').to_csv(text_filename, sep=' ', index=False, header=False)
    else:
        print(f"Warning: Source file {src} not found.")

filename_series = pd.Series(group_by_obj_train.groups.keys())

# Split the data into 80% for training and 20% for testing
train_filenames, test_filenames = train_test_split(filename_series, test_size=0.2, random_state=42)

# Move files and save data for the training set
for filename in train_filenames:
    save_data(filename, train_folder, group_by_obj_train)

# Move files and save data for the testing set
for filename in test_filenames:
    save_data(filename, test_folder, group_by_obj_test)
```
